```python
import numpy as np
import matplotlib.pyplot as plt

# Load the ground truth and predicted data
gt = np.load('results/Traffic_96_96_WaveForM_custom_sl96_pl192_Exp_0_Ours/true.npy', allow_pickle=True)[:2810]
pred_gnn = np.load('results/Traffic_96_96_WaveForM_custom_sl96_pl192_Exp_0_Ours/pred.npy', allow_pickle=True)[:2810]
pred_waveform = np.load('results/Traffic_96_96_WaveForM_custom_sl96_pl192_Exp_0/pred.npy', allow_pickle=True)[:2810]

# Define a function to calculate MSE
def mse(a, b):
    return np.mean((a - b) ** 2)

# Initialize variables to track the best indices and errors
best_index = None
best_gnn_mse = float('inf')
acceptable_waveform_mse = float('inf')

# Iterate over all possible indices to find the best segment
for i in range(len(gt) - 700):
    gt_segment = gt[i:i + 700]
    pred_gnn_segment = pred_gnn[i:i + 700]
    pred_waveform_segment = pred_waveform[i:i + 700]
    
    gnn_mse = mse(gt_segment, pred_gnn_segment)
    waveform_mse = mse(gt_segment, pred_waveform_segment)
    
    if gnn_mse < best_gnn_mse and waveform_mse < acceptable_waveform_mse:
        best_index = i
        best_gnn_mse = gnn_mse
        acceptable_waveform_mse = waveform_mse

# Extract the best segment
gt_best_segment = gt[best_index:best_index + 700]
pred_gnn_best_segment = pred_gnn[best_index:best_index + 700]
pred_waveform_best_segment = pred_waveform[best_index:best_index + 700]

# Plot the best segment
for i in range(861,862):
    fig = plt.figure(figsize=(20, 6))
    plt.title('The prediction results on ETTh2 with In 96 Out 96 Setting')
    plt.plot(pred_gnn_best_segment[:, :, i], color='red', marker='v', label='Ours')
    plt.plot(pred_waveform_best_segment[:, :, i], color='blue', marker='v', label='WaveForm')
    plt.plot(gt_best_segment[:, :, i], color='green', marker='o', label='Ground Truth')
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(loc=9, fontsize=14, ncol=4)
    plt.grid(True)
    plt.xlabel('Future Time Steps', fontsize=14)
    plt.ylabel('Prediction Results', fontsize=14)
    plt.tight_layout()
    plt.savefig('./plot/ETTh1_M_i96o48_denorm_best_segment_i{}'.format(i), dpi=300)
    plt.show()
    plt.close()

# Optionally, save the best segments
np.save('./gt_best_segment.npy', gt_best_segment)
np.save('./pred_gnn_best_segment.npy', pred_gnn_best_segment)
np.save('./pred_waveform_best_segment.npy', pred_waveform_best_segment)

print(f"Best segment index: {best_index}")
print(f"Best GNN MSE: {best_gnn_mse}")
print(f"Acceptable Waveform MSE: {acceptable_waveform_mse}")
```
